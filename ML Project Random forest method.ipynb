```python
import pandas as pd
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt
import sklearn

data=pd.read_csv("C:\\Users\\Ram\\Desktop\\heart.csv");
data
```

```python
import statsmodels.formula.api as smf
reg=smf.ols(formula="target ~ cp + thalach + slope + oldpeak + exang + ca",data=data).fit()
reg.summary()
```

```python
x=data.as_matrix(['cp','thalach','slope','oldpeak','exang','ca'])
y=data.target


from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python

classifier = RandomForestClassifier(n_estimators=15,criterion="entropy",max_depth=3,min_samples_split=10)
classifier.fit(x_train,y_train)



y_pred=classifier.predict(x_test)

res=classifier.score(x_test,y_test)
    
print(res)
```
